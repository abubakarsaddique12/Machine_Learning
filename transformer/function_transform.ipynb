```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import scipy.stats as stats
from sklearn.preprocessing import FunctionTransformer
from sklearn.model_selection import train_test_split
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
X=df.drop("species",axis=1)
y=df['species']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.kdeplot(X_train['petal_length'])
plt.title('Age PDF')

plt.subplot(122)
stats.probplot(X_train['petal_length'], dist="norm", plot=plt)
plt.title('Age QQ Plot')

plt.show()
```

```python
transform=FunctionTransformer(func=np.log1p)
X_train_transform=transform.fit_transform(X_train)
X_test_transform=transform.transform(X_test)
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.kdeplot(X_train_transform['petal_length'])
plt.title('Age PDF')

plt.subplot(122)
stats.probplot(X_train_transform['petal_length'], dist="norm", plot=plt)
plt.title('Age QQ Plot')

plt.show()
```
